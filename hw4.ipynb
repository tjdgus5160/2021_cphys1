```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import math
from scipy.misc import derivative
```

```python
def Error_Function(x):
    return (2/(np.pi)**(1/2))*(np.exp(-x**2))

def integrate_ErF(Error_Function,a,b,n):
    s, x, dx = -1, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*Error_Function(x+dx)+0.5*Error_Function(x))
        x = x + dx
    return s

print(integrate_ErF(Error_Function,0,1,1))
print(integrate_ErF(Error_Function,0,2,4))

x = np.arange(-10,10,0.01)
plt.axis([-10,10,-2,2])
plt.plot(x,[integrate_ErF(Error_Function,-10,k,300) for k in x])
plt.xlabel('x')
plt.ylabel('Error_Function')
```

```python
def Error_Function_Taylor(x, n):
    f = 0
    for i in range(n+1):
        f = f + (2/np.pi)*(((-1)**i)*(x**(2*i+1)))/((2*i+1)*math.factorial(i))
        i = i+1
    return f

plt.plot(x, Error_Function_Taylor(x, 1), color = 'blue')
plt.plot(x, Error_Function_Taylor(x, 10), color = 'red')
plt.plot(x, Error_Function_Taylor(x, 100), color = 'purple')
plt.plot(x,[integrate_ErF(Error_Function,-10,k,300) for k in x], color = 'green')
plt.axis([-10,10,-2,2])
plt.show()
```

```python

def integrate(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

def for_diff(f,x,dx):
    return (f(x+dx)-f(x))/dx
def back_diff(f,x,dx):
    return (f(x)-f(x-dx))/dx
def cen_diff(f,x,dx):
    return (f(x+dx/2)-f(x-dx/2))/dx

dx = 0.2
x = np.arange(-5,5,0.1)
diff1 = [for_diff(sin,i,dx) for i in x]
diff2 = [back_diff(sin,i,dx) for i in x]
diff3 = [cen_diff(sin,i,dx) for i in x]

plt.plot(x,diff1)
plt.plot(x,diff2)
plt.plot(x,diff3)
plt.plot(x,sin(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def M(w,k):
  j=np.zeros((40,40))
  m=1
  d=2*k-m*(w**2)
  z=d-k
  for i in range(40):
    if i == 0:
      j[i][i]=z
      j[i][i+1]=-k
    elif i == 39:
      j[i][i]=z
      j[i][i-1]=-k
    else:
      j[i][i]=d
      j[i][i-1]=-k
      j[i][i+1]=-k
  return j
```

```python
def b(c):
    k=np.zeros((40,1))
    k[0]=c
    return k
```

```python
def b1(c):
    k=np.zeros((40,1))
    k[0]=c
    k[39]=c
    return k
```

```python
def matrix_graph():
    
    y=np.linalg.solve(M(1,1),b(1))
    y1=np.linalg.solve(M(2,1),b(1))
    y2=np.linalg.solve(M(3,1),b(1))
    y3=np.linalg.solve(M(4,1),b(1))

    x=[i for i in range(40)]
    plt.plot(x,y,label='w=1')
    plt.plot(x,y1,label='w=2')
    plt.plot(x,y2,label='w=3')
    plt.plot(x,y3,label='w=4')
    plt.xlabel('i')
    plt.ylabel('ai')
    plt.legend()
    plt.show()
```

```python
matrix_graph()
```

```python
def matrix_two_graph():
    y=np.linalg.solve(M(1,1),b1(1))
    y1=np.linalg.solve(M(2,1),b1(1))
    y2=np.linalg.solve(M(3,1),b1(1))
    y3=np.linalg.solve(M(4,1),b1(1))

    x=[i for i in range(40)]
    plt.plot(x,y,label='w=1')
    plt.plot(x,y1,label='w=2')
    plt.plot(x,y2,label='w=3')
    plt.plot(x,y3,label='w=4')
    plt.xlabel('i')
    plt.ylabel('ai')
    plt.legend()
    plt.show()
```

```python
matrix_two_graph()
```

```python
w가 작을수록 d의 값이 커진다. d가 커짐에 따라 그 차이가 더 커지는 것을 확인할 수 있다.
한쪽 끝에서 힘을 줄때는 반대쪽 벽면의 힘이 0이되고 양쪽에서 힘을 가하면 중간에서 힘이 0이됨을 알 수 있다.
```

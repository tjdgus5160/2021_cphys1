```python
import numpy as np
import matplotlib.pyplot as plt

def Random_walk(n):
    x = 0
    y = 0
    time = [x]
    position = [y] 
 
    for i in range (1,n+1):
        move = np.random.uniform(0,1)
        if move < 0.5:    
            x += 1
            y += 1  
        if move > 0.5:    
            x += 1
            y += -1
        time.append(x)
        position.append(y)
    return [time,position]

Random_walk = Random_walk(10000)
plt.plot(Random_walk[0],Random_walk[1],'r-', label = "Random_walk")
plt.title("1-D Random Walks")
plt.show()
```

```python
def Random_walk(n):
    x = 0
    y = 0
    time = [x]
    position = [y] 
 
    for i in range (1,n+1):
        move = np.random.uniform(0,1)
        if move < 0.5:    
            x += 1
            y += 1  
        if move > 0.5:    
            x += 1
            y += -1
        time.append(x)
        position.append(y)
    return [time,position], position[-1]

N = 10000
n = 100
X = []
for i in range(N):
    X.append(Random_walk(n)[1])
    
plt.hist(X, bins=75, density = True, alpha = 1)
plt.show()
```

```python
import numpy as np
import sympy
from sympy import *
```

```python
x = symbols('x', cls=Function)
r = symbols('r')
t = symbols('t')
x(t)
```

```python
Eq(x(t).diff(t), r * x(t) - x(t)**2)
```

```python
deq1 = Eq(x(t).diff(t), -x(t) - x(t)**2)
psol1 = dsolve(deq1, ics={x(0):2})
psol1
```

```python
plot(psol1.rhs,(t,0,10))
```

```python
psol2 = dsolve(deq1, ics={x(0):-2})
psol2
```

```python
plot(psol2.rhs,(t,0.7,10))
```

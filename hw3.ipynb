```python
def Leibniz(a):
    f = 1
    for n in range(1,a):
        if n % 4 == 1:
            f += (1*(-1)**(n))/(n+2)
        elif n % 4 == 3:
            f += (1*(-1)**(n-1))/(n+2)
    return f
    
x = np.arange(0,100,2)
y = [Leibniz(1), Leibniz(3), Leibniz(5), Leibniz(7), Leibniz(9), Leibniz(11), Leibniz(13), Leibniz(15), Leibniz(17), Leibniz(19), Leibniz(21), Leibniz(23), Leibniz(25), Leibniz(27), Leibniz(29), Leibniz(31), Leibniz(33), Leibniz(35), Leibniz(37), Leibniz(39), Leibniz(41), Leibniz(43), Leibniz(45), Leibniz(47), Leibniz(49), Leibniz(51), Leibniz(53), Leibniz(55), Leibniz(57), Leibniz(59), Leibniz(61), Leibniz(63), Leibniz(65), Leibniz(67), Leibniz(69), Leibniz(71), Leibniz(73), Leibniz(75), Leibniz(77), Leibniz(79), Leibniz(81), Leibniz(83), Leibniz(85), Leibniz(87), Leibniz(89), Leibniz(91), Leibniz(93), Leibniz(95), Leibniz(97), Leibniz(99)]
plt.xlabel('n')
plt.plot(x,y,label='Leibniz formula for π')
plt.hlines(np.pi/4,0,100,color='orange',label='π/4')
plt.legend()
plt.show()
```

```python
def sinc(n):
    x = np.linspace(-10,10,n)
    y = np.sin(x)/x
    
    plt.plot(x, y)

sinc(100)
```

```python
def sinc_Taylor(x,n):
    f = 0
    for i in range(n+1):
        f = f + ((-1)**i * x**(2*i))/(math.factorial(2*i+1))
    return f

    
x = np.arange(-10,10,0.1)
plt.plot(x,sinc_Taylor(x,1), color = 'purple')
plt.plot(x,sinc_Taylor(x,4), color = 'blue')
plt.plot(x,sinc_Taylor(x,8), color = 'red')
plt.plot(x,sinc_Taylor(x,12), color = 'green')
plt.axis([-10,10,-1,1])
plt.show()
```

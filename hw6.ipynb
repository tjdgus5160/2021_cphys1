```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
g = 10
L = 10
state0 = [1.0, 0.0]
t = np.arange(0.0, 50.0, 0.01)

def f(state, t):
    theta, w = state 
    return w, -(g/L)*np.sin(theta)

states = odeint(f, state0, t)
plt.plot(t,states[:,0])
```

```python
g, L = 10, 10
dt=0.001
t = np.arange(0,50,dt)
theta , w = np.array([]), np.array([])
theta0, w0 = 1, 0

for i in t:
    theta0 = theta0 + dt*w0
    w0 = w0 - dt*g*theta0/L
    theta=np.append(theta,theta0)
    w=np.append(w,w0)

plt.plot(t,theta)
```

```python
g = 10
L = 10
beta = 0.1
A = 1
v = 1
state0 = [1.0, 0.0]
t = np.arange(0.0, 50.0, 0.01)

def f(state, t):
    theta, w = state 
    return w, A*np.cos(v*t)-beta*w-(g/L)*np.sin(theta)

states = odeint(f, state0, t)
plt.plot(t,states[:,0])
```
